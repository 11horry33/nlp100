[cabochaについて](https://yu-nix.com/archives/python-cabocha/)

1段目（*で始まる行）、左から順に
* 文節番号
* かかり先の文節番号
* 主辞/機能語の位置
* 係り関係のスコア

それ以降の行は、左から順に 　　　※表層系と品詞の間は"\t"、品詞以降は","で区切られている
* 表層形（元の文章上の表記）
* 品詞
* 品詞細分類1
* 品詞細分類2
* 品詞細分類3
* 活用形
* 活用型
* 原形
* 読み
* 発音

```python
#40. 係り受け解析結果の読み込み（形態素）

class Morph:
    def __init__(self, morph):
        surface, tmp = morph.split("\t")
        tmp = tmp.split(",")
        self.surface = surface
        self.base = tmp[6]
        self.pos = tmp[0]
        self.pos1 = tmp[1]

# １行ごとに読み込む
with open("ai.ja.txt.parsed") as f:
    text = f.read().split("\n")

morphs = []
sentences = []
for line in text[:-1]:
    if line[0] == "*":
        continue
    elif line == "EOS":
        sentences.append(morphs)
        morphs = []
    else:
        morphs.append(Morph(line))

# 冒頭の説明文を表示する
for morphs in sentences[7]:
    print(vars(morphs))
```

```python
#41. 係り受け解析結果の読み込み（文節・係り受け）

class Chunk:
    def __init__(self, morphs, dst):
        self.morphs = morphs
        self.dst = dst
        self.srcs = []

# 係り元を追加するためのクラス
class Sentence:
    def __init__(self, chunks):
        self.chunks = chunks
        for chunk_id, chunk in enumerate(self.chunks):
            if chunk.dst:
                self.chunks[chunk.dst].srcs.append(chunk_id)


# １行ごとに読み込む
with open("ai.ja.txt.parsed") as f:
    text = f.read().split("\n")

morphs = []
chunks = []
sentences = []
for line in text[:-1]:
    if line[0] == "*":
        if len(morphs) > 0:
            chunks.append(Chunk(morphs, dst))
            morphs = []
        dst = int(line.split(" ")[2][:-1])
    elif line == "EOS":
        chunks.append(Chunk(morphs, dst))
        sentences.append(Sentence(chunks))
        morphs = []
        chunks = []
        dst = None
    else:
        morphs.append(Morph(line))

for chunk in sentences[2].chunks:
    print(f"{''.join(morph.surface for morph in chunk.morphs)}\t{chunk.dst}\t{chunk.srcs}")
```

```python
#42. 係り元と係り先の文節の表示

for chunk in sentences[2].chunks:
    if int(chunk.dst) != -1:
        word1 = ''.join(morph.surface for morph in chunk.morphs if morph.pos != '記号')
        word2 = ''.join(morph.surface for morph in sentences[2].chunks[chunk.dst].morphs if morph.pos != '記号')
        print(f"{word1}\t{word2}")
```

```python
#43. 名詞を含む文節が動詞を含む文節に係るものを抽出

for chunk in sentences[2].chunks:
    if int(chunk.dst) != -1:
        word1 = ''.join(morph.surface for morph in chunk.morphs if morph.pos != '記号')
        word2 = ''.join(morph.surface for morph in sentences[2].chunks[chunk.dst].morphs if morph.pos != '記号')
        if "名詞" in [morph.pos for morph in chunk.morphs] and "動詞" in [morph.pos for morph in sentences[2].chunks[chunk.dst].morphs]:
            print(f"{word1}\t{word2}")
```

```python
#44. 係り受け木の可視化

# 有効グラフを作成する関数（引数：41で係り受け解析した文）
def tree_viz(sentence):
    graph = Digraph(format = "png")
    graph.attr('node', fontname = 'IPAGothic')
    for id, chunk in enumerate(sentence.chunks):
        word = ''.join(morph.surface for morph in chunk.morphs if morph.pos != '記号')
        graph.node(f"node_{id}", word) # ノードを追加
        if int(chunk.dst) != -1: # dst(係り受け先)が-1なら有効グラフは終点なので、エッジを追加しない
            graph.edge(f"node_{id}", f"node_{chunk.dst}") # エッジを追加
    return graph

graph = tree_viz(sentences[24])
graph.render('./nlp100_44.gv', view=True)
display_png(Image('./nlp100_44.gv.png'))
```

```python
#45. 動詞の格パターンの抽出

f = open('nlp100_45.txt', 'w')
for sentence in sentences:
    for chunk in sentence.chunks:
        for morph in chunk.morphs:
            if morph.pos == "動詞":
                particles = []
                for src in chunk.srcs:
                    particles.append(''.join(morph.surface for morph in sentence.chunks[src].morphs if morph.pos == '助詞'))
                if len(particles) > 0:
                    particles_sort = sorted(set(particles))
                    f.write(f"{morph.base}\t{' '.join(particles_sort)}\n")
                break
f.close()
```

```python
#46. 動詞の格フレーム情報の抽出

f = open('nlp100_46.txt', 'w')
for sentence in sentences:
    for chunk in sentence.chunks:
        for morph in chunk.morphs:
            if morph.pos == "動詞":
                particles = [] # 助詞
                words = []
                for src in chunk.srcs:
                    particle = ''.join(morph.surface for morph in sentence.chunks[src].morphs if morph.pos == '助詞')
                    if len(particle) > 0:
                        particles += particle
                        words.append(''.join(morph.surface for morph in sentence.chunks[src].morphs if morph.pos != '記号'))
                if len(particles) > 0:
                    particles_sort = sorted(set(particles))
                    f.write(f"{morph.base}\t{' '.join(particles_sort)}\t{' '.join(words)}\n")
                break
f.close()
```

```python
#47. 機能動詞構文のマイニング

f = open('nlp100_47.txt', 'w')
for sentence in sentences:
    for chunk in sentence.chunks:
        for morph in chunk.morphs:
            if morph.pos == "動詞":
                for i, src in enumerate(chunk.srcs):
                    # 係り元文節[srcs]の長さが２で「サ変接続 ＋ を」 
                    if len(sentence.chunks[src].morphs) == 2 and sentence.chunks[src].morphs[0].pos1 == "サ変接続" and sentence.chunks[src].morphs[1].surface == "を":
                        particles = [] # 述語に係る助詞
                        src_chunks = [] # 述語に係る文節                        
                        predicate = ''.join([sentence.chunks[src].morphs[0].base, sentence.chunks[src].morphs[1].base, morph.base]) # 述語
                        # その他の文節を探索
                        for other_chunk in chunk.srcs[:i] + chunk.srcs[i+1:]:
                            particle = [morph.surface for morph in sentence.chunks[other_chunk].morphs if morph.pos == '助詞']
                            # 助詞を含む文節があった場合、文節と助詞を取得
                            if len(particle) > 0:
                                particles += particle
                                src_chunks.append(''.join(morph.surface for morph in sentence.chunks[other_chunk].morphs if morph.pos != "記号"))
                        if predicate != None:
                            particles_sort = sorted(set(particles))
                            src_chunks_sort = sorted(src_chunks, key=lambda x : x[-1])
                            f.write(f"{predicate}\t{' '.join(particles_sort)}\t{' '.join(src_chunks_sort)}\n")
```

```python
#48. 名詞から根へのパスの抽出

def get_path1(sentence):
    for chunk in sentence.chunks:
        if "名詞" in [morph.pos for morph in chunk.morphs]:
            path = [''.join(morph.surface for morph in chunk.morphs if morph.pos != '記号')]
            while chunk.dst != -1:
                path.append(''.join(morph.surface for morph in sentence.chunks[chunk.dst].morphs if morph.pos != '記号'))
                chunk = sentence.chunks[chunk.dst]
            print(' -> '.join(path))

get_path1(sentences[2])
```

**解法**

名詞を含む文節 i, j について、以下の2パターン
* i →文節1→文節2→ j →根
 * 「 i →文節1→文節2→ j 」を表示
* i →文節1→ k →根かつ j →文節2→ k →根
 * 「 i →文節1｜ j →文節2｜ k 」を表示

※i, jの名詞部分はX, Yに変換

**組み合わせパターンの生成**

import itertools

comb = itertools.combinations('abc', 2)

list(comb)　→　[('a', 'b'), ('a', 'c'), ('b', 'c')]

```python
#49. 名詞間の係り受けパスの抽出

import itertools
import re

def get_path2(sentence):
    # 名詞を含む文節番号を抽出
    nouns = []
    for id, chunk in enumerate(sentence.chunks):
        if "名詞" in [morph.pos for morph in chunk.morphs]:
            nouns.append(id)
    # 名詞ペアを作成（組み合わせ）
    for i, j in itertools.combinations(nouns, 2):
        path_i = []
        path_j = []
        # i, j それぞれのパスを作成
        while i != j:
            if i < j:
                path_i.append(i)
                i = sentence.chunks[i].dst
            else:
                path_j.append(j)
                j = sentence.chunks[j].dst
        # 1つ目のパターン
        if len(path_j) == 0:
            chunk_X = "".join(morph.surface if morph.pos != "名詞" and morph.pos != "記号" else "X" for morph in sentence.chunks[path_i[0]].morphs)
            chunk_X = re.sub("X+", "X", chunk_X)
            chunk_Y = "".join(morph.surface if morph.pos != "名詞" and morph.pos != "記号" else "Y" for morph in sentence.chunks[i].morphs)
            chunk_Y = re.sub("Y+", "Y", chunk_Y)
            XtoY = [chunk_X] + ["".join(morph.surface for morph in sentence.chunks[n].morphs if morph.pos != "記号") for n in path_i[1:]] + [chunk_Y]
            print(" -> ".join(XtoY))
        # 2つ目のパターン
        else:
            chunk_X = "".join(morph.surface if morph.pos != "名詞" else "X" for morph in sentence.chunks[path_i[0]].morphs if morph.pos != "記号")
            chunk_X = re.sub("X+", "X", chunk_X)
            chunk_Y = "".join(morph.surface if morph.pos != "名詞" else "Y" for morph in sentence.chunks[path_j[0]].morphs if morph.pos != "記号")
            chunk_Y = re.sub("Y+", "Y", chunk_Y)
            chunk_k = ''.join([morph.surface for morph in sentence.chunks[i].morphs if morph.pos != "記号"])
            path_X = [chunk_X] + ["".join(morph.surface for morph in sentence.chunks[n].morphs if morph.pos != "記号") for n in path_i[1:]]
            path_Y = [chunk_Y] + ["".join(morph.surface for morph in sentence.chunks[n].morphs if morph.pos != "記号") for n in path_j[1:]]
            print(f'{" -> ".join(path_X)} | {" -> ".join(path_Y)} | {chunk_k}')

get_path2(sentences[2])
```
